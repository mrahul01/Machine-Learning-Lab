```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

df =pd.read_csv('Iris.csv')
df
```

```python
plt.figure(figsize=(10,6))
sns.pairplot(df,hue='Species',diag_kind='hist',palette='dark')
plt.show()
```

```python
X = df.drop(columns='Species')
y = df['Species']
```

## Logistic Regreesion

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=40)

LR_clf = LogisticRegression(solver='lbfgs',max_iter=1000)
LR_clf.fit(X_train,y_train)

LR_pred = LR_clf.predict(X_test)

LR_acc = accuracy_score(y_test,LR_pred)
LR_repo = classification_report(y_test,LR_pred)
print(f"The Accuracy of Logistic Regression is : {LR_acc}")
print(f"\nThe Classification report of Logistic Regression is :\n {LR_repo}")
```

## SVM

```python
from sklearn.svm import SVC
SVM_clf = SVC()
SVM_clf.fit(X_train,y_train)


SVM_pred = SVM_clf.predict(X_test)

SVM_acc = accuracy_score(y_test,SVM_pred)
SVM_repo = classification_report(y_test,SVM_pred)
print(f"The Accuracy of Support Vector Machine is : {SVM_acc}")
print(f"\nThe Classification report of Support Vector Machine is :\n {SVM_repo}")
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
DT_clf = DecisionTreeClassifier()
DT_clf.fit(X_train,y_train)


DT_pred = DT_clf.predict(X_test)

DT_acc = accuracy_score(y_test,DT_pred)
DT_repo = classification_report(y_test,DT_pred)
print(f"The Accuracy of Decision Tree is : {DT_acc}")
print(f"\nThe Classification report of Decision Tree is :\n {DT_repo}")
```

## Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
NB_clf = GaussianNB()
NB_clf.fit(X_train,y_train)


NB_pred = NB_clf.predict(X_test)

NB_acc = accuracy_score(y_test,NB_pred)
NB_repo = classification_report(y_test,NB_pred)
print(f"The Accuracy of Naive Bayes is : {NB_acc}")
print(f"\nThe Classification report of Naive Bayes is :\n {NB_repo}")
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
KNN_clf = KNeighborsClassifier(n_neighbors=2)
KNN_clf.fit(X_train,y_train)


KNN_pred = KNN_clf.predict(X_test)

KNN_acc = accuracy_score(y_test,KNN_pred)
KNN_repo = classification_report(y_test,KNN_pred)
print(f"The Accuracy of KNN is : {KNN_acc}")
print(f"\nThe Classification report of KNN is :\n {KNN_repo}")
```
