# 6. Prepare X (feature variables) and y (response variable)

```python
X = df[['TV']]
y = df['Sales']
```

# 7. Split Data into Training and Testing Sets (70%-Train 30%-Test) - X_train, y_train, X_test, y_test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Training set shape: {X_train.shape}, {y_train.shape}")
print(f"Testing set shape: {X_test.shape}, {y_test.shape}")
```

# 8. Print the shape of train data and test data

```python
print(f"Training data shape: {X_train.shape}")
print(f"Testing data shape: {X_test.shape}")
```

# 9. Perform Linear Regression (on train data) without using any library and print the coefficients values

```python
import numpy as np

X_train_with_intercept = np.c_[np.ones(X_train.shape[0]), X_train]

coefficients = np.linalg.inv(X_train_with_intercept.T.dot(X_train_with_intercept)).dot(X_train_with_intercept.T).dot(y_train)

print(f"Intercept: {coefficients[0]}")
print(f"Coefficient of X: {coefficients[1]}")
```

# 10.Create the model using the calculated coefficients and make predictions on the testing set

```python

y_pred = coefficients[0] + coefficients[1] * X_test

print("Predictions: ", y_pred)
```

# 11.Perform Linear Regression using library(sklearn)

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, y_train)

print(f"Intercept (sklearn): {model.intercept_}")
print(f"Coefficient of X (sklearn): {model.coef_}")
```

# 12.Print the intercept and coefficients

```python
print(f"Intercept (sklearn): {model.intercept_}")
print(f"Coefficient of X (sklearn): {model.coef_}")
```

# 13.Make predictions on the testing set - y_pred

```python

y_pred = model.predict(X_test)

print("Predicted values on test set: \n", y_pred)
```

# 14.Plot train data and the model

```python

plt.scatter(X_train, y_train, color='blue')
plt.plot(X_train, model.predict(X_train), color='red')
plt.title('Training Data and Linear Regression Model')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

# 15.Plot test data and the model

```python
# Plot test data and the model
plt.scatter(X_test, y_test, color='blue')
plt.plot(X_test, y_pred, color='red')
plt.title('Test Data and Linear Regression Model')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

# 16.Print Mean Squared Error and R Squared Value

```python
from sklearn.metrics import mean_squared_error, r2_score

# Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")

# Calculate R-squared
r2 = r2_score(y_test, y_pred)
print(f"R-squared: {r2}")
```

# 17.Calculate and print R Squared and Adjusted R Squared values without any library

```python
n = len(y_test)
p = X_train.shape[1]

# Calculate Adjusted R-squared
r2_adj = 1 - (1 - r2) * (n - 1) / (n - p - 1)
print(f"Adjusted R-squared: {r2_adj}")
```
